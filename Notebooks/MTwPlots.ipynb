```python
merge_table = pd.merge(stock, sector, on="Date", how="outer")
merge_table.head(50)
```

```python
merge_table.count()
```

```python
merge_table2=pd.merge(merge_table, friday, how="outer", on="Date")
merge_table2.head(100)
                
```

```python
merge_table2.count()
```

```python
merge_table2.columns
```

```python
merge_table2["Days of Week"]=dt.datetime.weekday
```

```python
merge_table2.head()
```

```python
merge_table2["Days of Week"]=pd.to_datetime(merge_table2.Date).dt.day_name()
```

```python
merge_table2["Year"]=pd.to_datetime(merge_table2.Date).dt.year
```

```python
merge_table2.head()
```

```python
merge_table2.head(50)
```

```python
merge_table2.count()#describe()
```

```python
Tickergroup=merge_table2.groupby(["Ticker","Year"])
```

```python
TickerMeans=pd.DataFrame(Tickergroup["Close"].mean()).reset_index()
```

```python
TickerMeans.head()
```

```python
PivotTickerMeans=TickerMeans.pivot(index="Year", columns="Ticker", values="Close")
```

```python
PivotTickerMeans.head(100)
```

```python
TickerSEM=Tickergroup["Close"].sem()
StdErrorTickerDF=pd.DataFrame(TickerSEM).reset_index()
StdErrorTickerDF.head()
```

```python
PivotCloseSEMs=StdErrorTickerDF.pivot(index="Year",columns="Ticker", values="Close")
PivotCloseSEMs.head()
```

```python
PivotCloseSEMs_df=pd.DataFrame(PivotCloseSEMs)
```

```python
np.arange(len(PivotTickerMeans))
```

```python
fig, ax=plt.subplots(figsize=(20,6))

for column in PivotTickerMeans.columns:
    ax.errorbar(np.arange(len(PivotTickerMeans)), PivotTickerMeans[column], yerr=PivotCloseSEMs_df[column], fmt="o", linestyle="dashed")

years=[2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
x_axis = np.arange(0, 20, 1)
labels=years
ax.set_ylabel("Close")
ax.set_xlabel("Year")
ax.set_xticks(x_axis)
ax.set_xticklabels(years)
ax.legend(loc="upper left")
ax.set_title("Close by Ticker Over Time")
ax.legend(PivotTickerMeans.columns)
```
