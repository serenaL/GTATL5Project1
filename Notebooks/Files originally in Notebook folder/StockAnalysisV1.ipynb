```python
#read in holidays data file
StockHoliday=pd.read_csv('../Stock Holidays.csv')
```

```python
for i in range(StockHoliday):
    StockHoliday["DateBefore"][i]= datetime.datetime.strptime(StockHoliday["Date"][i], "%m/%d/%y") + datetime.timedelta(days=-1)

```

```python
StockHoliday.head()
```

```python
ElevSectHoliday=pd.merge(ElevenSectors, StockHoliday, how="left", left_on="Date", right_on="Date")
```

```python
ElevSectHoliday.fillna(0)
```

```python
ElevSectHoliday['Weekday']=datetime.datetime.weekday(ElevSectHoliday["Date"])
```
